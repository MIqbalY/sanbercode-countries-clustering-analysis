```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

#code here
sns.scatterplot(data=df, x='carat', y='price', hue='clarity', size='depth', palette='ch:r=-.2,d=.3_r', edgecolor='none')
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
sns.set_style("ticks")
sns.catplot(data=df, x="clarity", y="price", kind='box', col='cut', hue='clarity', showfliers=False)
plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df_sample = df.sample(n=300, random_state=123)

#code here
sns.set_theme()
sns.pairplot(data=df_sample, hue='cut', markers=['p', 'o', 's', 'D', 'X'], corner=True, vars=['carat','depth','table','price'])
plt.show()
```

Expected Output:

![alt text](https://drive.google.com/uc?id=12rStINwUNGpqZd1iIGaqzTIvw9irAkJ0)

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
#kde
with sns.axes_style('white'):
  sns.jointplot(data=df_sample, x="carat", y="price", kind='kde')

#reg
sns.jointplot(data=df_sample, x="carat", y="price", kind='reg')
plt.xlim(0.25,2.25)
plt.show()
```
