<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

**Figure** merupakan objek yang mengandung elemen dari sebuah grafik, sedangkan **Axes** merupakan tempat dimana method untuk plotting didefinisikan, termasuk method untuk membuat dan memanipulasi garis *axis*, *grid*, ataupun *background*. Jika kita asumsikan seperti melukis, **Figure** adalah kertas atau kanvas, sedangkan **Axes** adalah gambarnya.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#eksplisit
fig = plt.figure()
ax = fig.add_subplot()

x = np.linspace(0, 20, 100)
y = np.cos(x/2)

ax.plot(x,y)

plt.show()
```

```python
#implicit
plt.plot(x,y)

plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)

ax.plot(x,y)
ax.plot(x,y2)
ax.plot(x,y3)

plt.style.use('seaborn')

fig.savefig('lineplot.png')

plt.show()
```
