Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
#code here
fig, ax = plt.subplots(facecolor='none', figsize=(18.5, 8), nrows=1, ncols=3, sharex=True, sharey=True)

x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
sin = np.sin(x)
cos = np.cos(x)

ax[0].plot(x, tan, color='blue', label='tanX')
ax[1].plot(x, sin, color='red', label='sinX')
ax[2].plot(x, cos, color='green', label='cosX')

tan_plot = ax[0].set_title('Grafik tanx')
sin_plot = ax[1].set_title('Grafik sinx')
cos_plot = ax[2].set_title('Grafik cosx')

fig.legend(loc='outside upper center')

plt.show()

# for i in range(3):
#     ax[i].text(0.5, 0.5, str((i)), fontsize=18, ha='center')
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(facecolor='none', figsize=(18,10))

x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)

ax.plot(x, y)
ax.set_title('Grafik CosX')

cos5 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
cos5.plot(x, y, color='darkblue')
cos5.plot(x, y5, color='#c2c20e')
cos5.set_title('Grafik CosX^5')

cos4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
cos4.plot(x, y, color='darkblue')
cos4.plot(x, y4, color='black')
cos4.set_title('Grafik CosX^4')

cos3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
cos3.plot(x, y, color='darkblue')
cos3.plot(x, y3, color='green')
cos3.set_title('Grafik CosX^3')

cos2 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
cos2.plot(x, y, color='darkblue')
cos2.plot(x, y2, color='red')
cos2.set_title('Grafik CosX^2')

plt.show()
```
