````python
import matplotlib.pyplot as plt

# Plot histogram
plt.hist(area, bins=50, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram of Area')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
````

Analisis bentuk sebaran data:

Dari histogram sebaran data yang telah dibuat, terlihat bahwa distribusi data area memiliki bentuk yang agak miring ke kanan (positively skewed) dengan ekor yang lebih panjang di sebelah kanan. Beberapa poin analisis yang dapat ditarik dari sebaran data ini adalah:

1. **Pusat Sebaran**: Mayoritas area berpusat di sekitar nilai 500 hingga 1500, dengan puncak frekuensi terletak di sekitar kisaran tersebut. Ini menunjukkan bahwa sebagian besar data berada di kisaran tersebut.

2. **Tinggi dan Frekuensi**: Frekuensi data cenderung menurun seiring dengan meningkatnya nilai area. Artinya, semakin besar nilai area, semakin sedikit sampel yang memiliki nilai tersebut.

3. **Outliers**: Terdapat beberapa nilai yang cukup jauh dari mayoritas distribusi, terutama di sebelah kanan histogram. Ini menunjukkan adanya outliers atau pencilan yang memiliki nilai area yang jauh lebih besar dari mayoritas sampel.

4. **Positively Skewed**: Distribusi data memiliki ekor yang lebih panjang di sebelah kanan dan memiliki kemiringan positif. Hal ini menunjukkan bahwa sebagian besar nilai area cenderung lebih kecil, namun terdapat beberapa nilai yang sangat besar yang menyebabkan ekor distribusi yang panjang ke kanan.

Dari analisis ini, kita dapat menyimpulkan bahwa sebagian besar sampel memiliki nilai area yang relatif kecil, namun terdapat beberapa sampel dengan nilai area yang jauh lebih besar, yang mungkin merupakan pencilan atau kasus-kasus khusus yang perlu diperhatikan secara lebih detil.
