# Soal 1. Pemahaman k-means clustering

Jelaskan bagaimana cara kerja k-means clustering!

Jawab disini:

**K Means clustering** adalah algoritma unsupervised machine learning yang digunakan untuk pengelompokan data dan pengenalan pola.

Caranya adalah dengan memilih beberapa titik data awal (k) secara acak, lalu memindah-mindahkannya hingga pengelompokan yang paling ideal ditemukan.

K Means clustering sering digunakan dalam berbagai bidang, seperti segmentasi gambar, segmentasi pelanggan, riset pasar, dan pengelompokan dokumen.

**Kelebihan K Means Clustering**

Berdasarkan Google, berikut ini adalah kelebihan K Means clustering:

* Sederhana: dapat digunakan oleh orang-orang dengan pengetahuan pemrograman yang terbatas.
* Mampu mengolah kumpulan data besar: dapat digunakan pada data dalam jumlah besar tanpa menghabiskan terlalu banyak waktu.
* Konvergensi: algoritma ini pada akhirnya akan menghasilkan kumpulan cluster yang stabil.
* Adaptasi dengan data baru: saat titik data baru tersedia, algoritma baru dapat dijalankan kembali untuk memasukkannya ke dalam cluster.
* Generalisasi: dapat melakukan generalisasi cluster dengan berbagai bentuk dan ukuran, seperti cluster elips, sehingga dapat digunakan untuk menemukan pola dalam berbagai kumpulan data.

**Kekurangan K Means Clustering**

Mengutip dari Google, kekurangan K Means clustering adalah:

* Memilih secara manual: jumlah cluster (k) harus dipilih secara manual, yang bisa jadi sulit dan subjektif. Memilih nilai k yang salah dapat mengakibatkan hasil pengelompokan yang buruk, dan memilih nilai k yang optimal membutuhkan pengetahuan yang cukup mengenai kumpulan data.
* Bergantung pada nilai centroid awal: karena bergantung pada nilai centroid yang ada di awal, algoritma ini mungkin tidak selalu dapat menemukan clustering yang ideal, sehingga perlu menjalankan algoritmanya beberapa kali untuk menemukan solusi optimal.
* Pengelompokan data dengan berbagai ukuran dan kepadatan: kelompok dalam data yang memiliki ukuran dan kepadatan berbeda dapat mengakibatkan beberapa cluster jadi tidak seimbang. Hal ini dapat mempersulit interpretasi cluster yang dihasilkan.
* Pengelompokan outlier: outlier adalah titik data yang secara signifikan berbeda dari titik data lain dalam suatu kumpulan data. Outlier dapat mendistorsi hasil pengelompokan dengan menarik centroid menjauh dari pusat cluster yang sebenarnya.
* Penghitungan data dengan banyak dimensi: dengan bertambahnya jumlah dimensi, pengelompokan data akan sulit karena perhitungan jarak antara titik data dan centroid menjadi lebih kompleks. Maka dari itu, diperlukan proses untuk mengurangi dimensi atau modifikasi algoritma.


**Cara kerja** algoritma K-Means dijelaskan dalam langkah-langkah di bawah ini:

Langkah-1: Pilih angka K untuk menentukan jumlah cluster.

Langkah-2: Pilih titik K atau centroid secara acak.

Langkah-3: Tetapkan setiap titik data ke centroid terdekat, yang akan membentuk cluster K yang telah ditentukan.

Langkah-4: Hitung varians dan tempatkan centroid baru dari setiap cluster.

Langkah-5: Ulangi langkah ketiga, yang berarti menetapkan kembali setiap titik data ke centroid terdekat baru dari setiap cluster.

Langkah-6: Jika ada penugasan ulang, lanjutkan ke langkah-4 jika tidak, lanjutkan ke FINISH.

Langkah-7: Model sudah siap




Algoritma ini didesain untuk memungkinkan kita mengelompokkan data ke dalam grup yang berbeda dengan cara yang lebih mudah berdasarkan variabel tertentu tanpa perlu melakukan proses training.

Hal ini karena k-means clustering merupakan algoritma unspervised learning berbasis centroid, dimana setiap cluster diasosiasikan dengan centroid. Tujuan utama dari algoritma ini adalah untuk meminimalkan jumlah jarak antara titik data dan cluster yang sesuai.

Algoritma K-means mengambil dataset yang tidak berlabel sebagai input, kemudian membagi dataset menjadi sejumlah k cluster, dan mengulangi proses tersebut sampai tidak menemukan cluster terbaik. Nilai k harus ditentukan sebelumnya dalam algoritma ini.
Algoritma k-means clustering melakukan dua tugas utama, yakni:

Menentukan nilai terbaik untuk titik pusat K atau centroid dengan proses iteratif (perulangan).
Menetapkan setiap titik data ke pusat k terdekat. Titik-titik data yang dekat dengan pusat-k tertentu, kemudian dibuatkan sebuah klusterOleh karena itu setiap cluster memiliki titik data dengan beberapa kesamaan, dan cukup jauh dari cluster lainnya.

# Soal 2. Clustering data menggunakan k-means clustering

Dalam soal ini, kalian diminta untuk melakukan clustering antara data Umur dan Skor Belanja(1-100). Namun, clustering tidak bisa dilakukan karena data memiliki outlier dan missing value. Maka, lakukanlah tahapan-tahapan berikut:

*   Handling missing value dengan menggunakan method ='ffill'
*   Handling outlier menggunakan metode Interquartile range
*   Lakukan rescaling data menggunakan StandardScaler
*   Lakukan clustering menggunakan kmeans clustering dengan k =2, dan visualisasikan..
*   Gunakan Elbow method untuk menemukan jumlah k kluster yang tepat
*   Gunakan k=3 (rekomendasi methode Elbow) lalu visualisasikan
*   Hitung silhoutte Coefficient dari ke dua hasil prediksi klustering tersebut










```python
df.info()
```

```python
df.fillna(method='ffill', inplace=True)
df.info()
```

```python
df_clean = df.copy()
```

```python
df_clean.describe()
```

```python
# Cek outlier
# Umur
# Pendapatan pertahun (juta)
# Skor Belanja (1-100)

sns.boxplot(x='Umur', data=df_clean)
plt.show()
```

```python
Q1 = df_clean['Umur'].quantile(0.25)
Q3 = df_clean['Umur'].quantile(0.75)
IQR = Q3 - Q1
batas_bawah = Q1-1.5*IQR
batas_atas = Q3+1.5*IQR
print(batas_bawah, batas_atas)
```

```python
df_clean_umur = df_clean[(df_clean['Umur'] > batas_bawah) & (df_clean['Umur'] < batas_atas)].copy()
```

```python
sns.boxplot(x='Pendapatan pertahun (juta)', data=df_clean)
plt.show()
```

```python
Q1 = df_clean_umur['Pendapatan pertahun (juta)'].quantile(0.25)
Q3 = df_clean_umur['Pendapatan pertahun (juta)'].quantile(0.75)
IQR = Q3 - Q1
batas_bawah = Q1-1.5*IQR
batas_atas = Q3+1.5*IQR
print(batas_bawah, batas_atas)
```

```python
df_clean_ppt = df_clean_umur[(df_clean_umur['Pendapatan pertahun (juta)'] > batas_bawah) & (df_clean_umur['Pendapatan pertahun (juta)'] < batas_atas)].copy()
```

```python
sns.boxplot(x='Skor Belanja (1-100)', data=df_clean)
plt.show()
```

```python
Q1 = df_clean_ppt['Skor Belanja (1-100)'].quantile(0.25)
Q3 = df_clean_ppt['Skor Belanja (1-100)'].quantile(0.75)
IQR = Q3 - Q1
batas_bawah = Q1-1.5*IQR
batas_atas = Q3+1.5*IQR
print(batas_bawah, batas_atas)
```

```python
df2 = df_clean_ppt[(df_clean_ppt['Skor Belanja (1-100)'] > batas_bawah) & (df_clean_ppt['Skor Belanja (1-100)'] < batas_atas)].copy()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20,5))

sns.boxplot(ax=axes[0], data=df2, x='Umur')
sns.boxplot(ax=axes[1], data=df2, x='Pendapatan pertahun (juta)')
sns.boxplot(ax=axes[2], data=df2, x='Skor Belanja (1-100)')

plt.show()
```

```python
df2.drop(['ID_pelanggan','Jenis Kelamin'], axis=1, inplace=True)
df2.describe()
```

```python
# Rescale
std = StandardScaler()
df_fit = std.fit_transform(df2.astype(float))

#clustering menggunakan k-means dengan k=2
kmeans = KMeans(n_clusters=2, n_init='auto').fit(df_fit)
labels_cluster = kmeans.labels_
centers1 = kmeans.cluster_centers_

df_std = pd.DataFrame(data=df_fit, columns=['Umur','Pendapatan pertahun (juta)','Skor Belanja (1-100)'])
df_std['cluster_2'] = labels_cluster

# visualisasi klaster
print("n_cluster = 2")
colors=[{"red","blue"}]
fig, axes = plt.subplots(figsize=(8,6))
# plt.scatter(df_std['Umur'], df_std['Skor Belanja (1-100)'], c=df_std['cluster'], edgecolor='green')
plt.scatter(df_std.Umur[df_std.cluster_2==0], df_std['Skor Belanja (1-100)'][df_std.cluster_2==0], c='blue', edgecolor='green')
plt.scatter(df_std.Umur[df_std.cluster_2==1], df_std['Skor Belanja (1-100)'][df_std.cluster_2==1], c='red', edgecolor='green')
plt.scatter(centers1[:, 0], centers1[:, 1], c='black', s=300)
plt.xlabel('Umur')
plt.ylabel('Skor Belanja (1-100)')
plt.show()

# elbow method untuk menentukan jumlah klaster yang tepat
sse = []
for k in range(1, 11):
    kmeans_elbow = KMeans(n_clusters=k, n_init='auto').fit(df_std)
    sse.append(kmeans_elbow.inertia_)
print("Elbow Method")
print("sse :")
print(sse)
plt.plot(range(1, 11), sse)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# clustering menggunakan k-means dengan k=3
kmeans2 = KMeans(n_clusters=3, n_init='auto').fit(df_std)
labels_cluster2 = kmeans2.labels_
centers2 = kmeans2.cluster_centers_

df_std2 = pd.DataFrame(data=df_fit, columns=['Umur','Pendapatan pertahun (juta)','Skor Belanja (1-100)'])
df_std2['cluster_3'] = labels_cluster2

# visualisasi klaster
print("n_cluster = 3")
fig, axes = plt.subplots(figsize=(8,6))
# plt.scatter(df_std2['Umur'], df_std2['Skor Belanja (1-100)'], c=df_std2['cluster'], edgecolor='green')
plt.scatter(df_std2.Umur[df_std2.cluster_3==0], df_std2['Skor Belanja (1-100)'][df_std2.cluster_3==0], c='blue', edgecolor='green')
plt.scatter(df_std2.Umur[df_std2.cluster_3==1], df_std2['Skor Belanja (1-100)'][df_std2.cluster_3==1], c='red', edgecolor='green')
plt.scatter(df_std2.Umur[df_std2.cluster_3==2], df_std2['Skor Belanja (1-100)'][df_std2.cluster_3==2], c='green', edgecolor='green')
plt.scatter(centers2[:, 0], centers2[:, 1], c='black', s=300)
plt.xlabel('Umur')
plt.ylabel('Skor Belanja (1-100)')
plt.show()

# hitung silhoutte coefficient
print("Silhoutte coefficient")
silhouette_coefficient = silhouette_score(df_std, kmeans.labels_)
print("k = 2 :",silhouette_coefficient)
silhouette_coefficient2 = silhouette_score(df_std2, kmeans2.labels_)
print("k = 3 :",silhouette_coefficient2)
```
