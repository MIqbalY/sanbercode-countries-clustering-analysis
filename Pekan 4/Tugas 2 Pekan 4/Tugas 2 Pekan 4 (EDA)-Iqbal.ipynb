# Soal 1: Pemahaman Teori
Jelaskan perbedaan antara Univariate, Bivariate dan Multivariate Analysis

Jawaban:

* **Univariate** analysis merupakan analisis deskriptif dengan satu variabel

* **Bivariate** analysis merupakan analisis relasi dengan dua variabel

* **Multivariate** analysis merupakan analisis yang menggunakan lebih dari atau sama dengan tiga variabel.

a. Univariate Analysis, merupakan analisis yang dilakukan untuk satu variabel atau pervariabel.
b.Bivariate Analysis, merupakan analisis yang dilakukan untuk menganalisis hubungan duavariabel.
c.Multivariate Analysis, merupakan analisis yang dilakukan untuk menganalisis hubunganlebih dari dua variabel.

# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#code here
df_cleaned = df.dropna()
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df2 = df_cleaned.groupby('species')[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].mean()
df2
```

# Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

*   Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
*   Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
*   Jelaskan insight yang diperoleh dari ke dua grafik tersebut.





```python
#code here
fig, ax = plt.subplots(facecolor='none', figsize=(10,6))
ax.pie(df2['flipper_length_mm'], labels=df2.index, autopct='%1.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(facecolor='none', figsize=(8,6))

plt.hist(df_cleaned['body_mass_g'], bins=10)
plt.xlabel('Body_mass_g')
plt.ylabel('Count')

plt.show()
```

Analisa:

Kedua grafik di atas adalah penerapan dari Univariate analysis yang merupakan analisis deskriptif dengan satu variabel.

Pada Pie chart variabel yang digunakan adalah panjang sirip atau flipper_length_mm. Dapat dilihat bahwa panjang sirip antara ketiga spesies penguin tersebut memiliki ukuran yang cukup serupa dengan selisih persentase yang kecil. Gentoo merupakan spesies dengan sirip paling panjang diikut oleh Chinstrap dan Adelie.

Pada Histogram variabel yang digunakan adalah berat badan atau body_mass_g. Dapat dilihat persebaran berat badan diantara ketiga jenis penguin. Jumlah penguin dengan berat badan sekitar 6000gr adalah yang paling sedikit sedangkan jumlah penguin dengan berat badan sekitar 3500-400gr adalah yang paling banyak.

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
#code here
sns.set_theme(style='ticks')
sns.catplot(x='sex', y='bill_length_mm', kind='bar', hue='sex', data=df_cleaned)
plt.show()
```

```python
# sns.set({'figure.figsize':(13,13)})
sns.set_theme(style='ticks', rc={'figure.figsize':(12,8)})
palette = "ch:s=-.2,r=.6"

sns.scatterplot(x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm', palette='ch:s=-.2,r=.6', data=df_cleaned)
plt.show()
```

Analisa:

Bivariate analysis merupakan analisis relasi dengan dua variabel

Pada Bar chart menampilkan data panjang paruh/bill_length_mm berdasarkan jenis kelamin. Dapat dilihat penguin jantan memiliki paruh lebih panjang daripada penguin betina dengan selisih panjang yang tidak begitu signifikan.

Pada Scatterplot umumnya digunakan untuk melihat apakah variabel memiliki hubungan yang berbanding lurus atau justru berbanding terbalik. Grafik di atas menunjukkan persebaran data perbandingan berat badan dan panjang sirip dengan pengelompokan berdasarkan panjang paruh. Dapat dilihat bahwa seiring bertambahnya panjang sirip dan berat badan, panjang paruh penguin juga semakin bertambah, Salah satu faktor yang mempengaruhi hal ini mungkin adalah usia penguin.

# Soal 5. Multivariate Analisis

*   Dengan menggunakan seaborn, buatlah sebuah pairplot keseluruhan data feature pada dataset (df) dengan argumen kind='bar', seperti expected di bawah:
*   Jelaskan hasil analisa dari multivariate yang dihasilkan



```python
#code here
sns.set_theme(style='ticks')
sns.pairplot(df)
plt.show()
```

Analisa:

Multivariate analysis merupakan analisis yang menggunakan lebih dari atau sama dengan tiga variabel.

Dengan pairplot kita dapat menganalisis langsung 4 variabel yaitu body_mass_g, flipper_length_mm, bill_length_mm, bill_depth_mm. Penggunaan bar chart dan scatterplot membantu kita untuk langsung melihat bagaimana jumlah dan persebaran data yang sedang dianalisis.
