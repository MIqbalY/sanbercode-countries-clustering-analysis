# Soal 1. Menentukan jenis data categorical atau numeric
Jelaskan data kategorikal (kualitatif) dan data numerik (kuantitatif)



Jawab disini:

**Data numerik** adalah data berwujud angka yang bisa didapat dari sebuah pengukuran. Misal, ukuran tinggi badan, berat badan, dan tekanan darah kita. Selain itu, data ini juga bisa diperoleh dari sebuah penghitungan, misalnya jumlah orang yang hadir di pesta pernikahan. Data numerik terbagi menjadi 2 yaitu, data diskrit dan data kontinu.

*   **Data kontinu** dapat direpresentasikan dalam berbagai nilai numerik, seperti bilangan desimal, bulat, dan lain-lain. Contohnya Suhu tubuh, Tinggi badan, Berat badan, dll.
*   **Data diskrit** merupakan data numerik yang hanya bisa direpresentasikan dengan bilangan bulat dan tidak dapat dibagi ke dalam unit yang lebih kecil. Contohnya Jumlah penumpang.

**Data kategorikal** merupakan data yang dapat dikelompokkan dan terbagi berdasarkan karakteristik atau ciri khasnya masing-masing. Dari data kategorikal, ada dua pembagian, yaitu nominal dan ordinal.

*   **Data nominal** adalah jenis pengelompokan data yang tidak memiliki keterkaitan dengan data lainnya dan tidak memiliki arti khusus. Jadi, data ini dapat dibedakan tanpa harus mengurutkan atau dibandingkan dengan data lainnya. Contohnya Jenis kendaraan, Warna iris mata, Jenis kelamin, dll.
*   **Data ordinal** adalah jenis pengelompokkan data yang memiliki urutan, atau harus disusun secara berurutan dengan mekanisme peringkat. Contohnya Sanbercode leaderboard, Rating produk/jasa/tempat, Level pendidikan, dll.

Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka:

ket:

- colormap adalah 'coolwarm'

- berikan size terhadap setiap data poin dengan nilai dari column WindAvgMPH, kalikan dengan 20 agar size terlihat lebih besar


b. Kemudian bandingkan data visualisasi diatas dengan visualisasi data (sample=1000) setelah handling missing value menggunakan:
- Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

```python
# tanpa handling missing value
fig, ax = plt.subplots(facecolor='none', figsize=(10.5,6))
sizes = df_sample['WindAvgMPH']*20

plt.scatter(x='HumidityAvgPercent', y='DewPointAvgF', c='TempAvgF', s=sizes, cmap='coolwarm', data=df_sample)
plt.xlabel("Humidity Average (%)")
plt.ylabel("Dew Point Average (F)")
plt.title("Austin Weather")
plt.show()
```

```python
df_sample.fillna(method='ffill', inplace=True)
```

```python
df_hmv=df_sample
```

```python
# handling missing value
fig, ax = plt.subplots(facecolor='none', figsize=(10.5,6))
sizes = df_hmv['WindAvgMPH']*20

plt.scatter(x='HumidityAvgPercent', y='DewPointAvgF', c='TempAvgF', s=sizes, cmap='coolwarm', data=df_hmv)
plt.xlabel("Humidity Average (%)")
plt.ylabel("Dew Point Average (F)")
plt.title("Austin Weather")
plt.show()
```

Analisa :

Berdasarkan kedua grafik di atas dapat dilihat penggunaan method ffill sebagai cara untuk menghandle missing value.

Fungsi Pandas dataframe.ffill() digunakan untuk mengisi nilai yang hilang dalam dataframe. 'ffill' adalah singkatan dari 'forward fill'. Baris yang terdapat missing value akan mendapat nilai baru dari baris yang berada di atasnya/baris sebelumnya.

Sehingga dapat dilihat beberapa titik baru yang menunjukkan data yang telah dihandle menggunakan method ffill.

```python
q1 = df2['House_Price'].quantile(0.25)
q3 = df2['House_Price'].quantile(0.75)
IQR = q3 - q1
batas_bawah = q1-(1.5*IQR)
batas_atas = q3+(1.5*IQR)
print(batas_bawah, batas_atas)
```

```python
df3 = df2[(df2['House_Price'] > batas_bawah) & (df2['House_Price'] < batas_atas)]
df3
```

```python
import seaborn as sns
sns.boxplot(x='House_Price', data=df3)
```
