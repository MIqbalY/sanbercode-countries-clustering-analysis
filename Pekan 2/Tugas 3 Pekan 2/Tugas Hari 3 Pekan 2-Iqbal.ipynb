<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

* **Pivoting** adalah suatu perubahan bentuk data yang memungkinkan kita untuk memutar data dalam DataFrame sehingga nilai-nilai unik dalam satu kolom dapat menjadi kolom baru dalam DataFrame yang terpisah
* **Melting** adalah suatu perubahan bentuk data yang mengubah tata letak DataFrame dari lebar menjadi panjang. Melting data memungkinkan kita untuk menggabungkan beberapa kolom menjadi satu kolom dan menyusun kembali data dalam format yang lebih terstruktur.

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)]
}

df_r1 = pd.DataFrame(data)
df_r1
```

Perhatikan data di atas, kemudian ubahlah bentuknya agar sesuai yang di harapkan.

```python
#code here
pivot_r1 = df_r1.pivot(index=['Tipe Ujian'],
                       columns=['Mata Pelajaran'],
                       values=['Rata2 Nilai Ujian','Rata2 Waktu Belajar (menit)'])
pivot_r1
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df_r2 = pd.DataFrame(data).set_index('Hari')
df_r2
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
df_r2 = df_r2.reset_index()
```

```python
#code here
melt_r2 = pd.melt(df_r2,
                  id_vars=['Hari'],
                  value_vars=['Kebun Binatang','Dufan','Disney','Bali'])
melt_r2
```
